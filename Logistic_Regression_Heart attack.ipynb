#### **Check for outliers and remove the outliers**

```python
def remove_outliers(data, columns):
    for column in columns:
        if column in data.columns:
            Q1 = data[column].quantile(0.25)
            Q3 = data[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data
```

```python
cleaned_data = remove_outliers(heart, ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca'])  #heart.columns
```

### **Machine Learning Process**

```python
X = heart.drop(columns = 'target')  #heart.drop('target', axis = 1)
y = heart['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 435)
```

#### **Apply the Logistic Regression on the data**

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)  #start the training process
```

```python
y_pred = log_reg.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

#### **Apply Decision Tree Classifier on the Data**

```python
dec_tree = DecisionTreeClassifier(max_depth = 9)
dec_tree.fit(X_train, y_train)
```

```python
dec_tree.get_depth()
```

```python
y_pred_2 = dec_tree.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_2)
```

#### **To check the overfitting**

```python
y_pred_train = dec_tree.predict(X_train)
```

```python
accuracy_score(y_train, y_pred_train)
```
